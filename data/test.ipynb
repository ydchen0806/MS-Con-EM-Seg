```python
import SimpleITK as sitk
import os
import numpy as np
from glob import glob
```

```python
nii_path = '/braindat/lab/chenyd/DATASET/unlabel_data_niigz'
nii_data = sorted(glob(os.path.join(nii_path, '*.nii.gz')))
print(len(nii_data))
```

```python
nii_dict = [
    {"image": image_name}
    for image_name in nii_data
]
```

```python
nii_dict
```

```python
# read nii
def read_nii(file_path):
    data = sitk.ReadImage(file_path)
    data = sitk.GetArrayFromImage(data)
    return data
```

```python
for i in nii_data:
    data = read_nii(i)
    print(data.shape)
    break
```
