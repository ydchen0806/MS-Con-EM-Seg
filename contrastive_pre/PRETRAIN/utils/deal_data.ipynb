```python
import numpy as np
data_total = glob(os.path.join('/h3cstore_ns/CT_data/CT_retrieval','CT*','ct','*.nii.gz'))
data_path = data['ct_path']
# search need data in data_total return list
need_data = data['PatientID'] +'_'+ data['ReportID']
need_data_path = []
for i in need_data:
    flag = 0
    for j in data_total:
        if i in j:
            need_data_path.append(j)
            flag = 1
            continue
    if flag == 0:
        need_data_path.append('None')
```

```python
need_data_path
```

```python
ct_path_need = [os.path.dirname(i) for i in need_data_path if i is not np.nan else np.nan]
data['ct_path'] = ct_path_need
data['path'] = need_data_path
data['path'].replace('None',np.nan,inplace=True)
data.to_csv('/h3cstore_ns/CT_data/CT_retrieval/meta/BIMCV_meta_cyd.csv',index=False)
```
